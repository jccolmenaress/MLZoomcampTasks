Questions 7: Value of the first element of w


```python
q7 = df[df['Make'] == 'Lotus']
q7
```

```python
q7 = q7[["Engine HP", "Engine Cylinders"]]
q7
```

```python
q7 = q7.drop_duplicates()
q7
```

```python
X = q7.to_numpy()
X
```

```python
XTX = np.dot(X.T,X )
XTX 
```

```python
XTX_inv = np.linalg.inv(XTX) 
XTX_inv
```

```python
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
```

```python
w = np.dot(np.dot(XTX_inv,X.T ),y)
```
